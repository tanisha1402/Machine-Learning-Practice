# Manuplulating a DataFrame

```python
cal_df.head()
```

```python
# adding a column in DataFrame
cal_df['Price'] = cal_dataset.target
```

```python
# dropping a row --> in terms of dropping a row , we should write df_name.drop(index=the row you wanna remove, axis = 0 (in case of rows, the axis should be 0))
# Imp : this row isn't drop permenantly , it's just shown as a dummy here.
# if you wanna permenantly drop a row , then you gotta make a sep new df for it
# cal_df1 = cal_df.drop(index=0,axis=0)
# like this
# this also will work -->
# cal_df.drop(index=0, axis=0, inplace=True)
cal_df.drop(index=0,axis=0)
```

```python
# Dropping multiple rows
# Temporary (not permanent)
cal_df.drop(index=[0, 2, 5])
# Permenant : cal_df.drop(index=[0, 2, 5], inplace=True)
```

```python
# dropping a column --> df_name.drop(colums='col name', axis = 1)
# Imp : this col isn't drop permenantly , it's just shown as a dummy here.
# if you wanna permenantly drop a col , then you gotta make a sep new df for it
# cal_df1 = cal_df.drop(columns='HouseAge', axis=1)
# like this
# this also will work -->
# cal_df.drop(columns='HouseAge', axis=1, inplace=True)
cal_df.drop(columns='HouseAge', axis=1)
# Alt : cal_df.drop('HouseAge', axis=1)
```

```python
# Dropping multiple columns
# Temporary (not permanent)
cal_df.drop(columns=['HouseAge', 'MedInc'])
# Alt : cal_df.drop(['HouseAge', 'MedInc'], axis=1)
# Permenant : cal_df.drop(columns=['HouseAge', 'MedInc'], inplace=True)
```

```python
# locating a row suing index value --> prints all details(cols data) of that index
cal_df.iloc[0]
```

```python
# locating a col using index
print(cal_df.iloc[:,0]) # first col
print(cal_df.iloc[:,1]) # second col
print(cal_df.iloc[:,2]) # third col
print(cal_df.iloc[:,-1]) # last col
```
