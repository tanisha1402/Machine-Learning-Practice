# Operators on Python
1. Arithmetic Operators
2. Asssignmnet Operators
3. Comparison Operators
4. Logical Operators
5. Identity Operators
6. Membership Operators

# 1. Arithmetic Operators

```python
 n1 = 20
 n2 = 10

 # addition
 sum = n1+n2
 print("Sum : ",sum)

 #substraction
 diff = n1-n2
 print("Difference : ",diff)

 #multiplication
 prod = n1*n2
 print("Product : ",prod)

 #division
 quot = n1/n2
 print("Quotient : ",quot)

 #modulus
 rem = n1%n2
 print("Remainder : ",rem)

 #exponent
 exp = n1**n2
 print("Exponent : ",exp)

```

# 3. Assignment Operators

```python
a = 10
b = 20
print(a==b)
print(a>b)
print(a<b)
print(a!=b)
print(a>=b)
print(a<=b)
```

# 4. Logical Operators
1. and
2. or
3. not

```python
a = 10
b = 20
print(a<b and b>a)
print(a<b or b>a)
print(not(a<b and b>a))
```

# 5. Identity Operators
1. is
2. is not

```python
a=5
b=5
print(a is b)
print(a is not b)
```

# 6. Membership Operator
1. in
2. in not

```python
l= [1,2,3,4,5]
print(3 in l)
print(3 not in l)
a = 9
print(a in l)
print(a not in l)
```
