# List Vs Numpy - Time Taken

```python
from time import process_time
```

Time taken by a list

```python
python_list =[i for i in range(500000)]
start_time = process_time()
python_list = [i+5 for i in python_list]
end_time= process_time()
print(end_time- start_time)
```

Time Taken By a Numpy Array

```python
nym_array = np.array([i for i in range(500000)])
start_time = process_time()
nym_array += 5
end_time = process_time()
print(end_time-start_time)
```

#  Initializing placeholders in np array

```python
# creating a np array of zeros
z = np.zeros((5,4))
print(z)
```

```python
# creating a np array of ones
o = np.ones((4,5))
print(o)
```

```python
# creating a np array with a particular value
f = np.full((4,5),6)
print(f)
```

```python
# creating an identity matrix
i = np.eye(5)
print(i)
```

```python
# creating a np array with random values
r = np.random.random((3,4))
print(r)
```

```python
# random integer values within specific range
ri = np.random.randint(10,100,(5,6))
print(ri)
```

```python
# array with even spaced values --> specifying the number of values required
r = np.linspace(10,20,7)
print(r)
```

```python
# array of evenly spaced values --> specifying the step
e = np.arange(10,20,5)
print(e)
```

```python
# converting a list to a np array
l = [1,2,3,4,5]
print(l)
n_arr = np.asarray(l)
print(n_arr)
type(n_arr)
```

# Mathematical Operations For NP array

```python
l1 = [1,2,3,4,5]
l2 = [5,4,3,2,1]
print(l1+l2) # it only concatenates , but doesn't perform addition on the values.
```

```python
a = np.random.randint(10,100,(3,4))
b = np.random.randint(10,100,(3,4))
print(a)
print(b)
```

```python
# actually performs mathematical operations on the arrays
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
# another way
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```
