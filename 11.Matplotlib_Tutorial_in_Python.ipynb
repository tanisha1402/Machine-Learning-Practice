#Matplotlib
useful for making plots

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# to get data for out plots
import numpy as np
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# generate a sine wave of x values(x-->the values themselves, y-->the sine values of x)
plt.plot(x,y)
plt.show()
```

```python
# generate a cosine wave of x values(x-->the values themselves, y-->the cosine values of x)
plt.plot(x,z)
plt.show()
```

```python
# adding titles, x axis & y axis label
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
plt.plot( x , y , 'r+')
plt.show()
```

```python
plt.plot(x,y,'g*')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

# Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
Language = ['English','Bangla','Latin','Japanese','German']
People = [100,60,78,67,99]
ax.bar(Language,People)
plt.xlabel('LANGUAGE')
plt.ylabel('PEOPLE')
plt.title('LANGUAGE VS PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','Bangla','Latin','Japanese','German']
people = [100,60,78,67,99]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
# Scatter Plot
x = np.linspace(1,10,30)
y = np.sin(x)
z = np.cos(x)
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'r')
ax.scatter(x,z,color='b')
plt.show()
```

```python
# 3d Scatter
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Greens')
```
