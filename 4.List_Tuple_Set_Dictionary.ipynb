# Mutable Objects
1. List
2. Set
3. Dictionary


```python
# lists should be included in square brackets
my_list1= [1,2,3,4,5]
print(my_list1)
type(my_list1)
```

```python
# lists can have multiple data types
my_list2=[1,3.9,"Hi World"]
print(my_list2)
```

lists are mutable --> changable

```python
# adding elements in list, use append func
my_list1.append("You can add elemets in lists, meaning it's mutable")
print(my_list1)
```

```python
# you can also print lists elements using index
print(my_list1[1])
print(my_list1[2:6])
```

```python
# lists allow duplicate values
my_list3= [1,2,3,1,2,"list allow duplicate values",5,5]
print(my_list3)
```

```python
print(len(my_list3))
```

```python
# initializing an empty lisy
my_list4= []
print(my_list4)
```

```python
my_list4.append(909)
print(my_list4)
```

```python
# deleting an item or element in list, use list_name.remove[index] or del list_name[index]
my_list5= [1,2,3,4,5,6,7,89,8]
my_list5.remove(1)
print(my_list5)
del my_list5[0]
print(my_list5)
```

```python
# join two lists
list_3 = [1,2,3,4,5]
list_4 = [6,7,8,9,10]

list_5 = list_3 + list_4
print(list_5)
```

# Tuple

```python
# tuples are included in first brackets
my_tuple1= (1,2,3,4)
print(my_tuple1)
```

```python
# tuples allow multiple data type
my_tuple2= (1,2,3.9,"Hi World")
print(my_tuple2)
```

```python
# converting list to tuple
list1 = [1,2,3,"Hi converting list to tuple"]
print(list1)
print(type(list1))
tuple1 = tuple(list1)
print(tuple1)
print(type(tuple1))
```

```python
# we can print an item or element in tuple using index, tuple_name[index]
print(tuple1[0])
print(tuple1[1:4])
```

# Tuples are immutable --> unchangable

```python
tuple1.append(2)
```

# Set

```python
# sets are included in curly brakets
set1 = {1,2,3}
print(set1)
```

```python
# you can't print an item or element in sets using index
print(set1[1])
```

```python
# set don't allow duplicate values
set2= {1,1,2,3,4,5,2,1,3,"set don't allow duplicate values, it wouldn't print or included it in itself"}
print(set2)
```

```python
# converting list into set
mylist = [1,2,3,4,5]
myset = set(mylist)
print(myset)
type(myset)
```

# Dictionary
1. key - value pair like set
2. it's included in curly brackets
3. doesn't allow any duplicate values

```python
my_dictionary= {'name':'Tanisha','age': 21, 'gender' : 'female'}
print(my_dictionary)
```

```python
# you can print an item or elements in dictionary using index
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['gender'])
```

```python
# dictionary doesn't allow duplicate keys
my_dictionary= {'name':'Tanisha','age': 21, 'gender' : 'female', 'name':'Tanisha','name2':'Tanisha'}
print(my_dictionary)
```
