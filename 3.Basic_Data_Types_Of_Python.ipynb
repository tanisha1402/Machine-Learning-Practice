# Basic data types of python
1. Interger
2. Floating Point
3. Complex
4. Boolean
5. String


```python
# integer
a = 10;
print(a)
type(a)
```

```python
# floating point
f = 1.88
print(f)
type(f)
```

```python
# complex numbers
c = 1+ 9j
print(c)
type(c)
```

# Convertion from one type to another

```python
# int to float
x = 10
print(x)
type(x)
```

```python
y = float(x)
print(y)
type(y)
```

```python
# float to int
x = 5.88
print(x)
type(x)
```

```python
y = int(x)
print(y)
type(y)
```

# Boolean
