# Seaborn Library
used for data visualization

```python
# importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting the theme
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

# Visualizing the data
data=tips  Uses the built-in tips dataset.

x='total_bill', y='tip'  Plots total bill on X-axis and tip on Y-axis.

col='time'  Creates separate plots for Lunch and Dinner.

hue='smoker'  Colors points by smoker/non-smoker.

style='smoker'  Changes point shape by smoker status.

size='size'  Scales point size by group size.

Overall  Shows relation between bill and tip across time, smoking status, and party size.

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
cal_housing = fetch_california_housing()
```

```python
print(cal_housing)
```

```python
cal_df = pd.DataFrame(data=cal_housing.data,columns=cal_housing.feature_names)
```

```python
cal_df['PRICE']=cal_housing.target
```

```python
print(cal_housing)
```

```python
cal_df.head()
```

```python
# Distribution Plot
sns.distplot(cal_df['PRICE'])
```

# Correlation :

1. Positive
2. Negative

```python
# correlation using heat map
correlation = cal_df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
